### **Utils**

```python
import matplotlib.pyplot as plt
```

```python
def get_max(list):
  max = 0
  for a in list:
    if a > max:
      max = a
  return max
```

```python
def read_files(path, file_names):
  data = []
  item = []
  for file_name in file_names:
    path = path + file_name
    inputs = open(path, "r")
    index = 0
    for input in inputs:
      test = input.split('\n\n')

      if "#" in test[0]:
        item = []
      elif "*" in test[0]:
        item.insert(len(item), test[0])
      elif test[0] != '\n':
        item.insert(len(item), test[0])
      else:
        data.insert(len(data), item)
        index += 1
    path = path.replace(file_name, "")
  return data
```

```python
def create_plotter():
  fig, ax = plt.subplots(figsize=(5,5))
  for index, geophone in enumerate(geophones):
    x = geophone[0]
    y = geophone[1]
    ax.scatter(x, y, marker = "^", s=200)
    ax.annotate('Geophone '+ str(index+1), (x, y))
  return fig, ax
```

### **Calibration**

```python
max_length = 1000
geophones = [(-50, 500), (500, 1050), (1050, 500), (500, -50)]

file_names = [
              "data_0_500.txt", #G1
              "data_500_1000.txt", #G2
              "data_1000_500.txt", #G3
              "data_500_0.txt" #G4
              ]

data = read_files(path, file_names)

max_amplitude = 0

for input in data:
  for c in input:
    chunks = c.split(" ")
    if chunks[1] == '1:' or chunks[1] == '2:' or chunks[1] == '3:' or chunks[1] == '4:':
      if int(chunks[2]) > max_amplitude:
        max_amplitude = int(chunks[2])
      if int(chunks[2]) == 541:
        print("asd", chunks[1])

print(max_amplitude)
lower_limit = 350
end_to_end = max_amplitude - lower_limit
```

### **Position Prediction using RSS (Recieved Signal Strength) Method**



```python
index = 0
predictions = []
real_locations = []
distances = []

for input in data:

  print("#", index + 1)

  real_x = int(input[0].split(" ")[1])
  real_y = int(input[1].split(" ")[1])
  index += 1
  
  sensor1 = []
  sensor2 = []
  sensor3 = []
  sensor4 = []

  for c in input:
    chunks = c.split(" ")
    if chunks[1] == '1:':
      sensor1.insert(len(sensor1), int(chunks[2]))
    elif chunks[1] == '2:':
      sensor2.insert(len(sensor2), int(chunks[2]))
    elif chunks[1] == '3:':
      sensor3.insert(len(sensor3), int(chunks[2]))
    elif chunks[1] == '4:':
      sensor4.insert(len(sensor4), int(chunks[2]))

  # ------------------------------------------
  #                  INPUT
  # ------------------------------------------

  # Testing Data
  geo1 = get_max(sensor1)
  print(geo1)
  geo2 = get_max(sensor2)
  print(geo2)
  geo3 = get_max(sensor3)
  print(geo3)
  geo4 = get_max(sensor4)
  print(geo4)
  print("\n")

  # ------------------------------------------
  #                CALCULATION
  # ------------------------------------------
  areas = [1, 2, 3, 4]
  area = 0

  predict_location = [0, 0]
  real_location = [real_x, real_y]
  real_locations.insert(len(real_locations), [real_x, real_y])

  geo13 = geo1 - geo3
  geo24 = geo2 - geo4
  print("geo13: ", geo13)
  print("geo24: ", geo24)

  if (geo13 < 0):
    areas.remove(1)
    geo13 = abs(geo13)
    predict_location[0] = max_length - (((end_to_end/2) - (geo13/2))/(end_to_end/2)) * (max_length/2)
  else:
    areas.remove(3)
    predict_location[0] = (((end_to_end/2) - (geo13/2))/(end_to_end/2)) * (max_length/2)

  if (geo24 < 0):
    areas.remove(2)
    geo24 = abs(geo24)
    predict_location[1] = (((end_to_end/2) - (geo24/2))/(end_to_end/2)) * (max_length/2)
  else:
    areas.remove(4)
    predict_location[1] = max_length - (((end_to_end/2) - (geo24/2))/(end_to_end/2)) * (max_length/2)

  if 1 in areas and 2 in areas:
    area = 1
  elif 2 in areas and 3 in areas:
    area = 2
  elif 3 in areas and 4 in areas:
    area = 3
  elif 4 in areas and 1 in areas:
    area = 4

  # ------------------------------------------
  #            EUCLIDEAN DISTANCE
  # ------------------------------------------
  import math
  distance = math.sqrt((real_x - predict_location[0])**2 + (real_y - predict_location[1])**2)
  print("Distance from real position: ", round(distance, 3), "cm")
  # print(str(round(distance, 3)).replace(".", ","))
  # if index % 10 == 0:
  #   print("real coordinate x: ", real_x, ", y: ", real_y)
  #   print("--------------------------------------------------------------------------------")
  #   print("\n")
  distances.insert(len(distances), distance)
  
  # ------------------------------------------
  #                  PLOTING
  # ------------------------------------------
  tx_square_side = max_length + 200
  rx_square_side = max_length + 200

  fig, ax = fig, ax = create_plotter()

  ax.scatter(real_location[0], real_location[1])
  ax.annotate('Real', (real_location[0], real_location[1]))

  ax.scatter(predict_location[0], predict_location[1])
  predict_label = 'Predict ' + "(" + str(predict_location[0]) + ", " + str(predict_location[1]) + ")"
  ax.annotate(predict_label, (predict_location[0], predict_location[1]))

  x_values = [real_x, predict_location[0]]
  y_values = [real_y, predict_location[1]]
  plt.plot(x_values, y_values)

  plt.show()

  prediction = [predict_location[0], predict_location[1]]
  predictions.insert(len(predictions), prediction)

  print("Durian was falling in area ", str(area))
  print("real coordinate x: ", real_x, ", y: ", real_y)
  print("prediction coordinate x: ", predict_location[0], ", y: ", predict_location[1])
  print("\n")
  print("--------------------------------------------------------------------------------")
  print("\n")
```
